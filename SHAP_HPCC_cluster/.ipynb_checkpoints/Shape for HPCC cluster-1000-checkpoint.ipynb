```python
N_sample = 5
df = pd.read_csv("full_dataframe_1node_2hrs_23dims.csv").drop('timestamp', 1).sample(N_sample, random_state=2020)
print(df.shape)
df.columns = [c.replace("192.168.13.108.443-","") for c in df.columns]
df.head()
```

```python
centers_df  = pd.read_csv("data_cluster.csv")
centers_df.head()
```

```python
common_cols = set(centers_df.columns).intersection(set(df.columns))
len(common_cols)
```

```python
centers_df = centers_df[list(common_cols) ]
centers_df
```

```python
def find_cluster(point):
    dist_list = []
    for i in range(centers_df.shape[0]):
        dist = np.linalg.norm(point-centers_df.iloc[i])
        dist_list.append(dist) 
        
    cluster = np.argmin(dist_list)
    return cluster

def find_cluster_matrix(matrix):
    res = []
    for point in matrix:
        res.append(find_cluster(point))
    return np.array(res)
```

```python
X_train, X_test = train_test_split(df, test_size = 0.3, random_state=2020)
X_train.shape, X_test.shape
```

```python
X_train
```

```python
X_train = np.array(X_train, dtype=float)
X_test = np.array(X_test, dtype=float)
```

```python
explainer = shap.KernelExplainer(find_cluster_matrix, X_train)
shap_values = explainer.shap_values(X_test)

shap_importance = pd.DataFrame({"feature_name":list( df.columns ), "shap_value": np.abs(shap_values[0]).sum(axis = 0)}).sort_values("shap_value", ascending=False)
shap_importance["shap_value"] = shap_importance["shap_value"] *100 / max(shap_importance["shap_value"])

shap_importance.to_csv(f"shap_variable_1000.csv")

## plot 
plt.figure(figsize=(8, 6))
sns.lineplot(data=shap_importance, x="feature_name" , y = "shap_value" )
```
